## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Create y_df with the Attrition and Department columns
y = attrition_df[['Attrition', 'Department']]
y
```

```python
# Create a list of at least 10 column names to use as X data

X = attrition_df[
    ['Education', 'Age', 'DistanceFromHome', 'JobSatisfaction', 'OverTime',
     'StockOptionLevel', 'WorkLifeBalance', 'YearsAtCompany',
     'YearsSinceLastPromotion', 'NumCompaniesWorked']
]

# Create X_df using your selected columns

X_df = X

# Show the data types for X_df

X.info()
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42,stratify=y)
```

```python
# Create a StandardScaler

scaler = StandardScaler()

# Fit the StandardScaler to the training data

scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)



# Fit the encoder to the training data

y_encoded = encoder.fit_transform(y[['Department']])


# Create two new variables by applying the encoder
# to the training and testing data

y_train_encoded = encoder.fit_transform(y_train[['Department']])
y_test_encoded = encoder.transform(y_test[['Department']])

print(y_train_encoded[:5])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

num_features = X_train.shape[1]
# Create the input layer
input_layer = Input(shape=(num_features,))

# Create at least two shared layers
shared_dense1 = Dense(64, activation='relu')
shared_dense2 = Dense(32, activation='relu')

x = shared_dense1(input_layer)
x = shared_dense2(x)

output_layer = Dense(1, activation='sigmoid')(x)

model = Model(inputs=input_layer, outputs=output_layer)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
num_department_features = y_train_encoded.shape[1]
department_input = Input(shape=(num_department_features,), name='department_input')

# Create the hidden layer
department_hidden = Dense(64, activation='relu')(department_input)

department_output = Dense(10, activation='softmax', name='department_output')(department_hidden)

department_model = Model(inputs=department_input, outputs=department_output)

department_model = Model(inputs=department_input, outputs=department_output)

department_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

department_model.summary()
# Create the output layer

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
num_department_features_1 = y_train_encoded_1.shape[1]
department_input_1 = Input(shape=(num_department_features,), name='department_input')

# Create the hidden layer
department_hidden_1 = Dense(64, activation='relu')(department_input_1)

department_output_1 = Dense(10, activation='softmax', name='department_output_1')(department_hidden_1)

department_model_1 = Model(inputs=department_input_1, outputs=department_output_1)

department_model_1 = Model(inputs=department_input_1, outputs=department_output_1)

department_model_1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

department_model_1.summary()

# Create the output layer

```

```python
# Create the model
input_layer = Input(shape=(10,), name='input')

# Compile the model

shared1 = Dense(64, activation='relu', name='shared1')(input_layer)
shared2 = Dense(128, activation='relu', name='shared2')(shared1)

department_hidden = Dense(32, activation='relu', name='department_hidden')(shared2)
department_output = Dense(3, activation='softmax', name='department_output')(department_hidden)

attrition_hidden = Dense(32, activation='relu', name='attrition_hidden')(shared2)
attrition_output = Dense(2, activation='softmax', name='attrition_output')(attrition_hidden)

model = Model(inputs=input_layer, outputs=[department_output, attrition_output], name='model')

model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'categorical_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})
# Summarize the model
model.summary()
```

```python
# Train the model

history = model.fit(
    X_train,
    {'department_output': y_train_encoded, 'attrition_output': y_train_encoded_1},
   epochs=100,
    batch_size=32,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data


results = model.evaluate(
    X_test,
    {'department_output': y_test_encoded, 'attrition_output': y_test_encoded_1}
)

# Print results directly as shown in your example output
print("[", end="")
print(f"{results[0]},", end=" ")
print(f"{results[1]},", end=" ")
print(f"{results[2]},", end=" ")
print("]", end="")
```
